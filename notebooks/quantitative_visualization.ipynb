```python
# Import necessary libraries
import sys
import os
import matplotlib.pyplot as plt

# Add the scripts directory to the Python path
sys.path.append('../scripts')  # Adjust this path if your directory structure is different

# Import functions from the script
from quantitative_analysis import load_and_prepare_data, calculate_technical_indicators, visualize_data

# List of CSV files to process
csv_files = [
    '../data/AAPL_historical_data.csv',
    '../data/AMZN_historical_data.csv',
    '../data/GOOG_historical_data.csv',
    '../data/META_historical_data.csv',
    '../data/MSFT_historical_data.csv',
    '../data/NVDA_historical_data.csv',
    '../data/TSLA_historical_data.csv'
]

# Iterate through each CSV file
for file_path in csv_files:
    print(f'Processing {file_path}')
    
    # Load and prepare data
    df = load_and_prepare_data(file_path)
    
    # Calculate technical indicators
    df = calculate_technical_indicators(df)
    
    # Create a new figure for each CSV file
    plt.figure(figsize=(12, 12))
    
    # Plot Close Prices and Moving Average
    plt.subplot(3, 1, 1)
    plt.plot(df['Close'], label='Close Price')
    plt.plot(df['SMA'], label='SMA')
    plt.title(f'Close Price and SMA for {os.path.basename(file_path)}')
    plt.legend()
    
    # Plot RSI
    plt.subplot(3, 1, 2)
    plt.plot(df['RSI'], label='RSI')
    plt.title(f'RSI for {os.path.basename(file_path)}')
    plt.legend()
    
    # Plot MACD
    plt.subplot(3, 1, 3)
    plt.plot(df['MACD'], label='MACD')
    plt.plot(df['MACD_SIGNAL'], label='MACD Signal')
    plt.bar(df.index, df['MACD_DIFF'], label='MACD Difference', color='gray', alpha=0.5)
    plt.title(f'MACD for {os.path.basename(file_path)}')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```
